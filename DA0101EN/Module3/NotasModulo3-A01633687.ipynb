**Analysis of Variance (ANOVA):**<br><br>
Find the correlation between different groups of a categorical variable<br>
It returns two values: <br>
F-test score: calculates the ratio of variation between the group's mean over the variation within each of the sample groups(score less than 1)<br>
P-value: Shows wether the obtained result is statistically significant.(larger than 0.05)<br>

How to apply ANOVA: <br>
df_anova = df[["column1", "column2"]]<br>
grouped_anova = df_anova.groupby(["column1"])<br>
anova_result = stats.**f_oneway**( grouped_anova.get_group("string1InColumn1")["column2"], grouped_anova.get_group("string2InColumn1")["column2"] )<br>

**Pearson Correlation:**<br>
Measure the strength of the correlation between two features<br><br>
Gives two values:<br>
Correlation coefficient:<br>
Close to +1: Large Positive Relationship<br>
Close to -1: Large Negative Relationship<br>
Close to 0:  No Relationship<br>
P-value:<br>
P-value < 0.001: Strong certainty in the result<br>
P-value < 0.05 : Moderate certainty in the result<br>
P-value < 0.1  : Weak certainty in the result<br>
P-value > 0.1  : No certainty in the result<br>

Strong Correlation:<br>
Correlation coeficient is close to 1 or -1 and<br>
P-value is less than 0.001<br>

<br>
How to apply it:<br>
pearson_coef, p_value = stats.**pearsonr**( df["column1"] , df["column2"] )<br>
